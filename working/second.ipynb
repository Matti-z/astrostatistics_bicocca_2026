```python
mu = 0.001
sigma0 = 2
sample = np.random.exponential( mu, int(1e5))

def funct(x, sigma, mu):
    return x**3 * np.exp(- (x**2 / (2*sigma**2))) * np.exp(-mu *x)


integral = np.mean(funct(sample, sigma0, mu))
print (integral)
```
