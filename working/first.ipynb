```python
import numpy as np
import matplotlib.pyplot as plt
import random
import plotly as pt
```

# def 

```python
def gen_3():
    x = np.zeros(3)
    x0 = random.randint(0,2)
    x[x0] = 1
    return x

def gen_N(n):
    x = np.zeros(n)
    x0 = random.randint(0, n-1)
    x [x0] = 1
    return x


def conserver(n = 3, t = False):
    x = gen_N(n)
    for i in range (n):
        if (x[i] == 1):
            if (t == True): print('conserver choice:', i+1)
            return i
        
def switcher(first_choice, the_goat, n, goat_door, t):
    
    last_choice = conserver(n, t)
    while (last_choice == first_choice or last_choice == goat_door):
        last_choice = conserver(n, t)
    return last_choice


def repeat(s_times, n, t = False):
    all_first = np.zeros(s_times)
    all_second = np.zeros(s_times)
    all_goats = np.zeros(s_times)
    all_cars = np.zeros(s_times)
    for i in range (s_times):
        car = conserver(n, t)
        first_choice = conserver(n , t)
        the_goat = np.zeros(n)
        goat_door = random.randint(0, n-1)
        the_goat[goat_door] == 1 
        while (the_goat[first_choice] == 1):
                goat_door = random.randint(0, n-1)
                the_goat[goat_door] = 1
        all_first[i] = first_choice
        all_goats[i] = goat_door
        all_cars[i] = car
        second_choice = switcher(t = t, first_choice = first_choice, the_goat = the_goat, n = n, goat_door = goat_door,)
        all_second[i] = second_choice
    return all_first, all_second, all_goats, all_cars
```

```python
n = 3
number_of_times = 100
f, s, g, c = repeat(number_of_times, n)
win_conserver = 0
win_switcher = 0
for i in range (number_of_times):
    if (f[i] == c[i]):
        win_conserver +=1
    if (s[i] == c[i]):
        win_switcher +=1
print(f'conserver wins: {win_conserver} / {number_of_times} = {win_conserver/number_of_times}')
print(f'switcher wins: {win_switcher} / {number_of_times} = {win_switcher/number_of_times}')
plt.hist(f, 100)
plt.hist(s, 1000)
plt.show()
```
