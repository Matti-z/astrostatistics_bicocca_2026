# def 

```python
def gen_3():
    x = np.zeros(3)
    x0 = random.randint(0,2)
    x[x0] = 1
    return x

def gen_N(n):
    x = np.zeros(n)
    x0 = random.randint(0, n-1)
    x [x0] = 1
    return x


def conserver(n = 3):
    x = gen_N(n)
    for i in range (n):
        if (x[i] == 1):
            # print('conserver choice:', i+1)
            return i
        
def switcher(first_choice, goat_door,  n):
    
    last_choice = conserver(n)
    while (last_choice == first_choice or last_choice == goat_door):
        last_choice = conserver(n)
    return last_choice

def newcomer(goat, first_choice, n):
    newcomer_choice = conserver(n)
    while (newcomer_choice == first_choice or newcomer_choice == goat):
        newcomer_choice = conserver(n)
    return newcomer_choice

def repeat(s_times, n, t = False):
    all_first = np.zeros(s_times)
    all_second = np.zeros(s_times)
    all_goats = np.zeros(s_times)
    all_cars = np.zeros(s_times)
    all_newcomers = np.zeros(s_times)
    for i in range (s_times):
        car = conserver(n)
        first_choice = conserver(n)
        the_goat = np.zeros(n)
        goat_door = random.randint(0, n-1)
        while (goat_door == first_choice or goat_door == car):
                goat_door = random.randint(0, n-1)
                the_goat[goat_door] = 1
        all_first[i] = first_choice
        all_goats[i] = goat_door
        all_cars[i] = car
        second_choice = switcher( first_choice = first_choice, n = n, goat_door = goat_door,)
        all_second[i] = second_choice
        newcomer_choice = newcomer(goat = goat_door, first_choice = first_choice, n = n)
        all_newcomers[i] = newcomer_choice
    return all_first, all_second, all_goats, all_cars, all_newcomers


```

```python
n = 3
number_of_times = int(1e6)
f, s, g, c, nc = repeat(number_of_times, n)
win_conserver = 0
win_switcher = 0
win_newcomers = 0
errors = 0
for i in range (number_of_times):
    if (f[i] == c[i]):
        win_conserver +=1
    if (s[i] == c[i]):
        win_switcher +=1
    if (g[i] == c[i]):
        errors +=1
    if (nc[i] == c[i]):
        win_newcomers +=1
print(f"conserver's wins: {win_conserver} / {number_of_times} = {win_conserver/number_of_times}")
print(f"switcher's wins: {win_switcher} / {number_of_times} = {win_switcher/number_of_times}")
print(f"newcomer's wins: {win_newcomers} / {number_of_times} = {win_newcomers/number_of_times}")
print(f'il programma è andato male: {errors} volte capra presentatore e macchina coincidono')
# plt.hist(f, 100)
# plt.hist(s, 1000)
# plt.hist(c, 100)
# plt.show()
```
