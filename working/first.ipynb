```python

def gen_N(n):
    x = np.zeros(n)
    x0 = random.randint(0, n-1)
    x [x0] = 1
    return x


def conserver(n = 3):
    x = gen_N(n)
    for i in range (n):
        if (x[i] == 1):
            # print('conserver choice:', i+1)
            return i
        
def switcher(first_choice, goat_door,  n):
    
    last_choice = random.randint(0, n-1)
    while (last_choice == first_choice or last_choice == goat_door):
        last_choice = random.randint(0, n-1)
    return last_choice

def newcomer(goat, n):
    newcomer_choice = random.randint(0, n-1)
    while (newcomer_choice == goat):
        newcomer_choice = random.randint(0, n-1)
    return newcomer_choice

def goat(first_choice, car, n):
    # the_goat = np.zeros(n)
    goat_door = random.randint(0, n-1)
    while (goat_door == first_choice or goat_door == car):
        goat_door = random.randint(0, n-1)
        # the_goat[goat_door] = 1
    return goat_door

def no_value_in_list(array, value):
    array = np.array(array)
    array_no_value = array[array != value]
    return array_no_value.tolist()


#estraggo le porte delle capre, il problema è che non riesco a confrontare il vettore
def p_goats(first_choice, car, n, p):
    vector = np.linspace(0, n-1, n)
    vector = no_value_in_list(vector, first_choice)
    print('no first choice vector', vector, first_choice)
    vector = no_value_in_list(vector, car)
    if (first_choice == car):
        goats = np.random.choice(vector, p, replace = False)
        return goats
    print('no car vector', vector, car)
    goats = int(np.random.choice(vector, p, replace = False))
    print('goat, vector', goats, vector)
    #now vector has all the possible good choices for the others to choose on
    # print(vector)
    return goats

#scelta per gli switchers
def ultimate_swift_choice(first_choice, goats, n):
    t = 0
    t = switcher(first_choice = first_choice, goat_door = goats[0], n = n)
    for i in range(len(goats)):
        while (t == goats[i]):
            for j in range(len(goats)):
                while (t == goats[j]):
                    t = switcher(first_choice, goat, n)
    return t


#scelta per i newcomers
def ultimate_choice(goat_doors, n, p):
    t = 0
    t = newcomer(goat = goat_doors[0], n = n)
    for i in range(p):
        while (t == goat_doors[i]):
            for j in range (p):
                while (t == goat_doors[j]):
                    t = newcomer(goat=goat_doors[0], n = n)
    return t        



def repeat(s_times, n, p, t = False):
    all_first = np.zeros(s_times)
    all_second = np.zeros(s_times)
    all_goats = np.zeros(s_times)
    all_cars = np.zeros(s_times)
    all_newcomers = np.zeros(s_times)
    if (p>n): 
        print("errore!! non puoi aprire più porte di quante ce ne siano")
        return False
    if t == False:
        for i in range (s_times):
            car = random.randint(0, n-1)
            first_choice = random.randint(0, n-1)
            goat_door = goat(first_choice, car, n)
            # goat_door = random.randint(0, 0, n-1)
            # while (goat_door == first_choice or goat_door == car):
            #     goat_door = random.randint(0, 0, n-1)
            all_first[i] = first_choice
            all_goats[i] = goat_door
            all_cars[i] = car
            second_choice = switcher( first_choice = first_choice, n = n, goat_door = goat_door,)
            all_second[i] = second_choice
            newcomer_choice = newcomer(goat = goat_door, n = n)
            all_newcomers[i] = newcomer_choice
        return all_first, all_second, all_goats, all_cars, all_newcomers
    if t == True:
        for i in range(s_times):
            car = random.randint(0, n-1)
            all_cars[i] = car
            first_choice = random.randint(0, n-1)
            all_first[i] = first_choice
            other_goats = p_goats(first_choice = first_choice, car = car, n = n, p = p)
            possible_good_choices = np.linspace(0, n-1, n)
            for j in range (p):
                possible_good_choices = no_value_in_list(possible_good_choices, other_goats[j])
            # all_goats = goat_door #for recording all goats position i would need a matrix, for the moment no
            final_choice = np.random.choice(possible_good_choices, 1, replace = False)
            all_second[i] = final_choice[0] #this is the choice after p doors are opened
            newcomers_choice = np.random.choice(possible_good_choices, 1, replace = False) 
            all_newcomers[i] = newcomers_choice 
        # print(all_first, all_second)

        return all_first, all_second,  all_cars, all_newcomers


```

# n = 3 & p = 1

```python
n = 3
number_of_times = int(1e6)
p = 0
f, s, g, c, nc = repeat(number_of_times, n, p)
win_conserver = 0
win_switcher = 0
win_newcomers = 0
errors = 0
for i in range (number_of_times):
    if (f[i] == c[i]):
        win_conserver += 1
    if (s[i] == c[i]):
        win_switcher += 1
    if (g[i] == c[i]):
        errors += 1
    if (nc[i] == c[i]):
        win_newcomers += 1
print(f"conserver's wins: {win_conserver} / {number_of_times} = {win_conserver/number_of_times}")
print(f"switcher's wins: {win_switcher} / {number_of_times} = {win_switcher/number_of_times}")
print(f"newcomer's wins: {win_newcomers} / {number_of_times} = {win_newcomers/number_of_times}")
print(f'il programma è andato male: {errors} volte capra-presentatore e macchina coincidono')
# plt.hist(f, 100)
# plt.hist(s, 1000)
# plt.hist(c, 100)
# plt.show()
```
